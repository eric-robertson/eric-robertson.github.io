# Data Normalization
Since the data is quite skewed, i will have to take steps to un-skew the data. Lets look at how it is skewed and what we can do to alleviate the bias that creates.

```python
# Break down by tags
tag_representation =  pd.read_csv( "./youtube_data/tag_view.csv" )
tag_representation = tag_representation[['date', 'tag', 'weight', 'vid']].set_index('date')
tag_representation.index = pd.to_datetime(tag_representation.index)
tag_representation
```

```python
# See how the weight per-day is skewed forward
_ = tag_representation.groupby('date').sum()
fig, ax = plt.subplots()
plt.plot(_.index, _.weight)
ax.set(title='Weight over time shows increase', xlabel='date', ylabel='weight of tags')
fig.set_size_inches(21,11)
```

Above it is clear that the weight of the tags are rising with time, in part because we have more samples for closer dates. To alleviate this we can divide by the number of videos published each day to help remove the bias from more videos being produced. In addition, the number of videos published from ealier dates is so few that many days lack video data at all. To help with this, we can fill any null date with the most recently published video statistics to full gaps in the data and resuce the spikeyness we see.

```python
# Utility to find missing dates
def fill_empty_dates (df):
    print(df.index.min())
    dates = pd.date_range(df.index.min(), df.index.max())
    return df.reindex(dates)
```

```python
# See how the weight per-day is skewed forward
_sum = tag_representation.groupby('date').sum()
_count = tag_representation.groupby('date').count()

# Divide by number of videos
_ = pd.DataFrame({ "vids": _count['vid'], "weight": _sum['weight']})
_ ['ave'] = _['weight'] / _['vids']

# Fill empty dates
_unfilled = fill_empty_dates(_)
_filled = fill_empty_dates(_).fillna(method='ffill')

_filled.to_csv('./youtube_data/daily_weight.csv')

_filled
```

```python
# Plot Again
fig, ax = plt.subplots()
plt.plot(_filled.index, _filled.ave)
ax.set(title='Weight over time shows increase', xlabel='date', ylabel='weight of tags')
fig.set_size_inches(21,11)
```

Plot is more normalized and shows less bias in data, but there is clearly a lot of heavy spikes. I wonder why? After diving into the dataset more, these major spikes are mainly atributed to days where there was only one video listed and that video went viral. This leads to the chart above where a hyper successful video overestimates the average performance of a single day skewing the results. To fix this we can smooth the data further by allowing one day to blead over into the next. Below is one such method that helps with that.

```python
def shift(xs, n):
    e = np.empty_like(xs)
    if n >= 0:
        e[:n] = 0
        e[n:] = xs[:-n]
    else:
        e[n:] = 0
        e[:n] = xs[-n:]
    return e
```

```python
# Functions to help smooth the spikes we see in the data above

# Smooth out spikes into more rounded hills to help neighboring days that are lacking viral sensations
def dual_rolling_smooth ( data, d = 5 ):
    return scale ( 
        data + shift(data, 1)/2 + shift(data, 2)/3 + shift(data, 3)/4+ shift(data, 4)/5+ shift(data, 5)/6
    )
# Scale all the data so it is between 0 and 1
def scale ( data ):
    return (data - data.min()) / ( data.max() - data.min() )
```

```python
# Example to show smoothing with randomly generated spikey data
sparse_data = np.random.choice(3, 300, p=[0.95, 0.025, 0.025])

sparse_scaled = scale( sparse_data )
sparse_smoothed = dual_rolling_smooth ( sparse_data, d=3 )

fig, ax = plt.subplots()
plt.plot(sparse_scaled, label = 'scaled')
plt.plot(sparse_smoothed, label = 'smoothed', linewidth=2)
ax.set(title='What smoothing looks like', xlabel='x', ylabel='y')
ax.legend()
fig.set_size_inches(21,11)
```

```python
# Using the unfilled data from above as filling dates is a form of smoothing. The new method should replace it
_ = _unfilled.fillna(0)
_['smooth_2'] = dual_rolling_smooth( _['ave'], d=2 )
_['smooth_5'] = dual_rolling_smooth( _['ave'], d=5 )

fig, ax = plt.subplots()
plt.plot(_.index, _.smooth_2, label='2-day smooth')
plt.plot(_.index, _.smooth_5, label='5-day smooth')
ax.set(title='Weight over time shows increase', xlabel='date', ylabel='weight of tags')
fig.set_size_inches(21,11)
ax.legend()
```

Now that we have normalized the data to the best of our ability we can look more into what we see here. Though not as pronouced as before, there is still a clearly shows an increase in the average tag over time. This can be many things, an increase in interest in personal finance, or perhapse just a growth in youtube viewership in general. Furthermore, even with the smoothing we did, the few massive viral videos dominate the weights of the entire dataset, but that does make sense. The weight here is ment to represent how influential that video was as its a combined metric of the various engagement statistics. Those massive viral videos have huge weight to them and its justified given their numbers.
