```python
import requests
import json
import os
import pandas as pd
import datetime
import numpy as np
import matplotlib.pylab as plt
import matplotlib.dates as mdates
from IPython.display import JSON
import ast
import scipy.stats as stats
import source.util as util
```

```python
target_stocks = ['SPY', 'TSLA', 'AAPL', 'F', 'FB', 'NIO', 'NFLX', 'UBER', 'HYLN', 'SQ', 'AMZN', 'LMND', 'RCL', 'DAL']
def process_stocks () :
    
    prices = []
    index_loaded = False
    index = ''

    for stock in target_stocks:
        # Taking only the daily data from the last few years
        frame = pd.DataFrame(util.load_json( f'./stock_data/{stock}.json' )['Time Series (Daily)']).T 
        frame = frame[ frame.index > '2017-01-01']

        if not index_loaded:
            index_loaded = True
            index = frame.index

        # Taking only the close price
        frame = frame['5. adjusted close'].values
        prices.append( frame )

    clean_stocks = pd.DataFrame(prices).T
    clean_stocks.columns = target_stocks
    clean_stocks.index = index
    clean_stocks.to_csv( f"./stock_data/clean.csv" )
    
    return clean_stocks
```

```python
process_stocks()
```

```python
# From SKEW
def fill_empty_dates (df):
    dates = pd.date_range(df.index.min(), df.index.max())
    return df.reindex(dates)
def clamp_dates (df, start='2017-01-01', end='2020-12-05'):
    dates = pd.date_range(start, end)
    return df.reindex(dates)
def dual_rolling_smooth ( data, d = 5 ):
    return scale ( 
        pd.Series(data)
            .rolling(d).mean()[::-1]
            .rolling(d).mean()[::-1]
            .rolling(d*2).mean()[::-1]
            .rolling(d*2).mean()[::-1]
    )
# Scale all the data so it is between 0 and 1
def scale ( data ):
    return (data - data.min()) / ( data.max() - data.min() )
```

```python
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)

_ = dual_rolling_smooth( stocks['SPY'] )
_.index = pd.to_datetime(_.index)
plt.plot(_)
```
