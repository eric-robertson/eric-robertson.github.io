## Video Data

```python
last_file = 0
data_dump = './data/youtube'
```

```python
def load_video ( video ):
    global last_file
    
    url = f"https://www.googleapis.com/youtube/v3/videos?id={video}&key={youtube_key}&part=snippet,statistics"
    response = requests.get( url )
    json_body = response.text
    
    last_file = last_file + 1
    
    with open(f"{data_dump}/{last_file}.json", 'w') as outfile:
        outfile.write(json_body)
        
```

```python
def load_channel ( channelID, page  ):
    url = f"https://www.googleapis.com/youtube/v3/search?key={youtube_key}&channelId={channelID}&part=snippet,id&order=date&maxResults=20&pageToken={page}"
    response = requests.get( url )
    json_body = response.text
    print( response )
    
    print( json_body )
```

```python
def search_channel ( channelName):
    url = f"https://www.googleapis.com/youtube/v3/search?part=id%2Csnippet&q={channelName}&type=channel&key={youtube_key}"
    response = requests.get( url )
    json_body = response.text
    
    print( json_body)
    
    _id = json.loads(json_body)['items'][0]['id']
    
    return _id
```
