```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.pyplot import figure
```

```python
sparse_data = np.random.choice(3, 300, p=[0.95, 0.025, 0.025])
sparse_data
```

```python
plt.plot( sparse_data )
```

```python
def rolling_smooth ( data, d = 5 ):
    return scale( pd.Series(data).rolling(d).mean() )
def dual_rolling_smooth ( data, d = 5 ):
    return scale ( pd.Series(data).rolling(d).mean()[::-1].rolling(d).mean()[::-1] )
def scale ( data ):
    return (data - data.min()) / ( data.max() - data.min() )
```

```python
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)

plt.plot( rolling_smooth(sparse_data, d = 10))
plt.plot( scale(sparse_data) )
```

```python
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)

plt.plot( dual_rolling_smooth(sparse_data, d = 20))
plt.plot( scale(sparse_data) )
```
