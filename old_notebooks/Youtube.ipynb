## Processing the video results

```python
import requests, json, os
```

```python
channels = [ a for a in os.listdir('./data/youtube') if a[0] != '.' ]
channels
```

```python
def load_video_ids ( channel ):
    with open(f"./data/youtube/{channel}/_videos.txt", 'r') as vids:
        return list(map(lambda v : v[:11], vids.readlines()))
```

```python
load_video_ids(channels[0]) [0]
```

```python
def pull_video ( videoID ):
    url = f"https://www.googleapis.com/youtube/v3/videos?part=snippet,statistics&id={videoID}&key=AIzaSyAFpH3wvd_J8YdlJIEAFMwGyfStudVPXhE"
    response = requests.get( url )
    content = response.text.encode("ascii", "ignore").decode("utf-8")
    return content
```

```python
def process_channel ( channelId ):
    if not os.path.exists(f"./data/youtube/{channelId}/videos"):
        os.mkdir(f"./data/youtube/{channelId}/videos")
    videoIds = load_video_ids( channelId )
    for i,vid in enumerate(videoIds) :
        if ( len(vid) != 11 ):
            continue
        if ( os.path.exists(f"./data/youtube/{channelId}/videos/{vid}.txt")):
            print(f"vid allready processed {vid} {i}/{len(videoIds)}", end='\r')
            continue
        videoData = pull_video ( vid )
        with open (f"./data/youtube/{channelId}/videos/{vid}.txt", 'w') as out:
            out.write( videoData )
        print( f"saved video {vid} {i}/{len(videoIds)}", end='\r')
    print(f"Done with {len(videoIds)} videos from { channelId}                   ")
        
            
```

```python
for channel in channels:
    process_channel(channel)
```
